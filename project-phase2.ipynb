```python
import pandas as pd


activity_path = '/kaggle/input/datasets/arashnic/fitbit/mturkfitbit_export_4.12.16-5.12.16/Fitabase Data 4.12.16-5.12.16/dailyActivity_merged.csv'
sleep_path = '/kaggle/input/datasets/arashnic/fitbit/mturkfitbit_export_4.12.16-5.12.16/Fitabase Data 4.12.16-5.12.16/sleepDay_merged.csv'

activity_df = pd.read_csv(activity_path)
sleep_df = pd.read_csv(sleep_path)


print(" Top 5 Rows ")
display(activity_df.head())


print("\nStatistical Analysis ")
display(activity_df.describe())
```

```python

print("Missing Values:")
display(activity_df.isnull().sum())

display(sleep_df.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# histogram

sns.set_theme(style="whitegrid")


plt.figure(figsize=(10, 5))
sns.histplot(activity_df['TotalSteps'], bins=30, kde=True, color='purple')
plt.title('Distribution of Total Daily Steps')
plt.xlabel('Total Steps')
plt.ylabel('Frequency (Number of Days)')
plt.show()
```

The distribution of total daily steps is slightly right-skewed. The most frequent step count (the peak) is around 10,000 steps, which aligns with common fitness goals. However, there is a noticeable spike at exactly 0 steps, which indicates days where users likely did not wear their fitness trackers.

```python
#  histogram to see the distribution of Sleep
plt.figure(figsize=(10, 5))
sns.histplot(sleep_df['TotalMinutesAsleep'], bins=20, kde=True, color='dodgerblue')
plt.title('Distribution of Total Minutes Asleep')
plt.xlabel('Total Minutes Asleep')
plt.ylabel('Frequency (Number of Records)')
plt.show()
```

```python
# scatter plot to see the relationship between Steps and Calories
plt.figure(figsize=(10, 5))
sns.scatterplot(data=activity_df, x='TotalSteps', y='Calories', alpha=0.7, color='teal')
plt.title('Total Steps vs. Calories Burned')
plt.xlabel('Total Steps')
plt.ylabel('Calories Burned')
plt.show()
```

```python

active_mins = [
    activity_df['VeryActiveMinutes'].mean(),
    activity_df['FairlyActiveMinutes'].mean(),
    activity_df['LightlyActiveMinutes'].mean(),
    activity_df['SedentaryMinutes'].mean()
]
labels = ['Very Active', 'Fairly Active', 'Lightly Active', 'Sedentary (Sitting)']
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

# Create a Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(active_mins, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Average Distribution of Daily Activity')
plt.show()
```

```python

columns_to_check = ['TotalSteps', 'TotalDistance', 'VeryActiveMinutes', 'SedentaryMinutes', 'Calories']
subset_df = activity_df[columns_to_check]

# Correlation Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(subset_df.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Activity Metrics')
plt.show()
```

The heatmap reveals a perfect correlation (1.00) between 'TotalSteps' and 'TotalDistance'. This means they give the machine learning model the exact same information. Additionally, there is a strong positive correlation (0.62) between 'VeryActiveMinutes' and 'Calories' burned.

```python
#  Boxplot to check for outliers in Total Steps
plt.figure(figsize=(10, 3))
sns.boxplot(x=activity_df['TotalSteps'], color='orange')
plt.title('Boxplot of Total Steps (Checking for Outliers)')
plt.xlabel('Total Steps')
plt.show()
```

```python
# 1. Dropping redundant and mostly empty columns
activity_df = activity_df.drop(['TrackerDistance', 'LoggedActivitiesDistance'], axis=1)

# Keeping rows where the user actually took steps 
activity_df = activity_df[activity_df['TotalSteps'] > 0]

print("Remaining Columns in Activity Data:")
print(activity_df.columns)
```

I dropped the 'TrackerDistance' feature because the correlation heatmap revealed it is nearly identical to 'TotalDistance', making it redundant. I also dropped 'LoggedActivitiesDistance' because the statistical analysis showed that the vast majority of its values are 0.0, meaning it does not provide useful variance for the machine learning model

```python
# Converting Date columns from text to proper DateTime objects
activity_df['ActivityDate'] = pd.to_datetime(activity_df['ActivityDate'])
sleep_df['SleepDay'] = pd.to_datetime(sleep_df['SleepDay'])


print("Activity Date Type:", activity_df['ActivityDate'].dtype)
print("Sleep Date Type:", sleep_df['SleepDay'].dtype)
```

The 'ActivityDate' and 'SleepDay' columns were originally stored as string objects. I converted them into standard Pandas DateTime objects.

```python
# Checking for and dropping any exact duplicate rows in our data
print(f"Duplicates in Activity Data: {activity_df.duplicated().sum()}")
print(f"Duplicates in Sleep Data: {sleep_df.duplicated().sum()}")

activity_df = activity_df.drop_duplicates()
sleep_df = sleep_df.drop_duplicates()
```

I checked for and removed any duplicate records in both datasets. Leaving exact duplicate rows in the dataset can cause a machine learning model to overfit and give biased results.

```python

activity_df = activity_df.rename(columns={'ActivityDate': 'Date'})
sleep_df = sleep_df.rename(columns={'SleepDay': 'Date'})

#  Merge the datasets together based on User Id and the Date
merged_df = pd.merge(activity_df, sleep_df, on=['Id', 'Date'], how='inner')

display(merged_df.head())
```

I renamed the date columns to a standard 'Date' format and performed an 'inner join' merge on the 'Id' and 'Date' columns. This ensures that we only keep records where a user tracked both their physical activity and their sleep on the exact same day, which is essential for training an accurate classification model.This intentionally reduced the dataset. This is a deliberate quality-control choice.
